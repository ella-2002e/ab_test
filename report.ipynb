```python
from Bandit import *
```

```python
Bandit_Reward = [1, 2, 3, 4]
NumberOfTrials = 20000
```

```python
#Reporting EpsilonGreedy
result = EpsilonGreedy(Bandit).experiment(Bandit_Reward, NumberOfTrials, t=1)
EpsilonGreedy(Bandit).report(NumberOfTrials, result, algorithm = "EpsilonGreedy")
```

```python
#visualizing epsilongreedy
Visualization().plot1(NumberOfTrials, result, 'Epsilon Greedy')
```

```python
result_thompson = ThompsonSampling(Bandit).experiment(Bandit_Reward, NumberOfTrials)
```

```python
ThompsonSampling(Bandit).report(NumberOfTrials, result_thompson, 'ThompsonSampling')
```

```python
#visualizing learning curve of Thompson
Visualization().plot1( NumberOfTrials,result_thompson, algorithm ='Thompson Sampling')
```

```python
#comparing thompson and epsilon_greedy
Visualization().plot2(result, result_thompson)
```

```python
#comparing using win rate convergence
comparison(NumberOfTrials, result, result_thompson)
```
